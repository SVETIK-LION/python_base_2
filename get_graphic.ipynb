```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# use set_position
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

def fx(x):
    return -12 * (x ** 4) - 18 * (x ** 3) + 5 * (x ** 2) + 10 * x - 30

points_x = np.linspace(-100, 100, 100)
points_y = [fx(x) for x in points_x]

idx = np.argmax(points_y)
max_y = points_y[idx]
max_x = points_x[idx]

print('Корней данная функция не имеет')
print(f'Функция возврастает на промежутке (-inf, {max_y}]')
print(f'Функция убывает на промежутке [{max_y}, +inf)')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.fill_between(points_x, points_y, where=(points_x < max_x), color='g')
plt.fill_between(points_x, points_y, where=(points_x >= max_x), color='r')
plt.plot(points_x, points_y)
plt.plot(max_x, max_y, marker='o')
plt.show()
```
